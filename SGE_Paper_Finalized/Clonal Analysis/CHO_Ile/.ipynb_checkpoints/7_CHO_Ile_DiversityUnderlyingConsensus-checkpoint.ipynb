```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os

# ---------- Sample relabeling ----------
sample_label_map = {
    "Counts_BS28194A": "PreSel",
    "Counts_BS28016A": "10-H1",
    "Counts_BS28025A": "5-C2",
    "Counts_BS28033A": "5-D2",
    "Counts_BS28038A": "2.5-G1",
    "Counts_BS28196A": "2.5-G2",
    "Counts_BS28045A": "2.5-G8",
    "Counts_BS28046A": "2.5-H1",

}

subplot_title_map = {
    "ilvA mut 2__PGK-MTS": "PGK-MTS ilvA-L481F",
    "ilvG__Ef1a-MTS": "Ef1a-MTS ilvG",
    "ilvM__PGK-MTS": "PGK-MTS ilvM",
    "ilvB__PGK-MTS": "PGK-MTS ilvB",
    "ilvC__PGK-MTS": "PGK-MTS ilvC",
    "ilvD__PGK-MTS": "PGK-MTS ilvD",
}

# ---------- Load data ----------
def load_data(file_paths):
    all_data = []
    for file_path in file_paths:
        sample_name = os.path.basename(file_path).split('.')[0]
        try:
            df = pd.read_csv(
                file_path,
                sep=",",
                header=None,
                names=["Barcode_Pair", "Gene_Promoter", "Barcode", "Count"],
                dtype=str,
                skiprows=10,
                encoding='utf-8'
            )
        except Exception as e:
            print(f"Failed to load {file_path}: {e}")
            continue

        if df.shape[1] != 4:
            print(f"Unexpected columns in {file_path}: {df.columns}")
            continue

        df = df[df["Barcode_Pair"].str.contains("__")]
        df["Count"] = pd.to_numeric(df["Count"], errors="coerce")
        df["Sample"] = sample_label_map.get(sample_name, sample_name)

        total = df["Count"].sum()
        if total > 0:
            df["Count"] = df["Count"] / total * 100

        all_data.append(df)

    return pd.concat(all_data, ignore_index=True) if all_data else pd.DataFrame()

# ---------- Plot barcode usage heatmap ----------
def plot_multiple_barcode_heatmaps(df, gene_promoter_pairs, cmap="Reds"):
    pivot_dfs = []
    all_values = []

    for gp in gene_promoter_pairs:
        sub_df = df[df["Gene_Promoter"] == gp].copy()
        sub_df[["Gene_Barcode", "Promoter_Barcode"]] = sub_df["Barcode_Pair"].str.split("__", expand=True)
        top_gene_bcs = sub_df.groupby("Gene_Barcode")["Count"].sum().nlargest(3).index.tolist()
        top_prom_bcs = sub_df.groupby("Promoter_Barcode")["Count"].sum().nlargest(3).index.tolist()
        expected_barcode_pairs = [f"{g}__{p}" for g in top_gene_bcs for p in top_prom_bcs]
        sub_df["Barcode_Pair"] = sub_df["Gene_Barcode"] + "__" + sub_df["Promoter_Barcode"]
        sub_df = sub_df[sub_df["Barcode_Pair"].isin(expected_barcode_pairs)]

        pivot_df = sub_df.pivot_table(
            index="Sample",
            columns="Barcode_Pair",
            values="Count",
            aggfunc="sum",
            fill_value=0
        )
        for bc in expected_barcode_pairs:
            if bc not in pivot_df.columns:
                pivot_df[bc] = 0

        pivot_df = pivot_df.sort_index()
        pivot_df = pivot_df[sorted(pivot_df.columns)]
        pivot_dfs.append((gp, pivot_df))
        all_values.append(pivot_df.values)

    global_data = np.concatenate(all_values, axis=None)
    vmin, vmax = global_data.min(), global_data.max()

    num_plots = len(pivot_dfs)
    rows, cols = 2, 3
    fig, axes = plt.subplots(rows, cols, figsize=(cols * 2.2, len(pivot_dfs[0][1].index) * 0.25 * rows), squeeze=False)

    
    for idx, (gp, pivot_df) in enumerate(pivot_dfs):
        row = idx // cols
        col = idx % cols
        ax = axes[row, col]

        sns.heatmap(
            pivot_df,
            ax=ax,
            cmap=cmap,
            cbar=False,
            vmin=vmin,
            vmax=vmax
        )

        if col == 0:
            ax.set_yticks(np.arange(len(pivot_df.index)) + 0.5)
            ax.set_yticklabels(pivot_df.index.tolist(), rotation=0)
        else:
            ax.set_yticks([])
            ax.set_yticklabels([])

        # Relabel x-axis as 1, 2, 3, ...
        num_cols = pivot_df.shape[1]
        ax.set_xticks(np.arange(num_cols) + 0.5)
        ax.set_xticklabels([str(i + 1) for i in range(num_cols)], rotation=0)

        # Remove tick marks on both axes
        ax.tick_params(axis='both', length=0)

        ax.set_title(subplot_title_map.get(gp, gp), fontsize=14)
        ax.set_xlabel("")
        ax.set_ylabel("")

    fig.tight_layout(rect=[0, 0, 0.9, 1])
    
    # Get positions of all subplot axes
    all_pos = [ax.get_position() for row_axes in axes for ax in row_axes]
    y0 = min(pos.y0 for pos in all_pos)
    y1 = max(pos.y1 for pos in all_pos)
    height = y1 - y0
    
    # Centered vertical colorbar
    cbar_ax = fig.add_axes([0.92, y0, 0.015, height])

    sm = plt.cm.ScalarMappable(cmap=cmap, norm=plt.Normalize(vmin=vmin, vmax=vmax))
    sm.set_array([])
    cbar = fig.colorbar(sm, cax=cbar_ax)
    cbar.set_label("Percentage of Reads", fontsize=12)

    fig.text(0.5, -0.01, "BC Combination", ha='center', va='center', fontsize=12)
    fig.text(0.001, 0.5, "Clone", ha='center', va='center', rotation='vertical', fontsize=12)

    fig.savefig("/Users/trollj01/PycharmProjects/pythonProject1/20250527_SGE_FinalScripts/CHO_Isoleucine/CHO_Isoleucine_SummaryOutputs/CHO_Ile_Consensus_BCDiv.svg", format="svg", bbox_inches="tight")
    plt.show()

file_paths = [
'/Users/trollj01/PycharmProjects/pythonProject1/20250527_SGE_FinalScripts/Controls/Controls_Counts/output_csvs/Counts_BS28194A', #CHO Pre-sel
'/Users/trollj01/PycharmProjects/pythonProject1/20250527_SGE_FinalScripts/CHO_Isoleucine/CHO_Isoleucine_Counts/output_csvs/Counts_BS28016A',
'/Users/trollj01/PycharmProjects/pythonProject1/20250527_SGE_FinalScripts/CHO_Isoleucine/CHO_Isoleucine_Counts/output_csvs/Counts_BS28025A',
'/Users/trollj01/PycharmProjects/pythonProject1/20250527_SGE_FinalScripts/CHO_Isoleucine/CHO_Isoleucine_Counts/output_csvs/Counts_BS28033A',
'/Users/trollj01/PycharmProjects/pythonProject1/20250527_SGE_FinalScripts/CHO_Isoleucine/CHO_Isoleucine_Counts/output_csvs/Counts_BS28038A',
'/Users/trollj01/PycharmProjects/pythonProject1/20250527_SGE_FinalScripts/CHO_Isoleucine/CHO_Isoleucine_Counts/output_csvs/Counts_BS28196A',
'/Users/trollj01/PycharmProjects/pythonProject1/20250527_SGE_FinalScripts/CHO_Isoleucine/CHO_Isoleucine_Counts/output_csvs/Counts_BS28045A',
'/Users/trollj01/PycharmProjects/pythonProject1/20250527_SGE_FinalScripts/CHO_Isoleucine/CHO_Isoleucine_Counts/output_csvs/Counts_BS28046A',
]

# ---------- Run ----------
data = load_data(file_paths)

plot_multiple_barcode_heatmaps(
    data,
    gene_promoter_pairs=["ilvA mut 2__PGK-MTS", "ilvG__Ef1a-MTS", "ilvM__PGK-MTS", "ilvB__PGK-MTS", "ilvC__PGK-MTS", "ilvD__PGK-MTS"]
)
```
